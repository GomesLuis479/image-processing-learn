```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.image as mpimg
```

```python
# read the image

image_original = mpimg.imread(r'../images/dogs.jpg')
image = image_original.copy() # else it is read only

# display original

plt.imshow(image_original, cmap='gray')
```

```python
# converting to grayscale: avg method

R = image[:,:, 0]
G = image[:,:, 1]
B = image[:,:, 2]

grayscale = (R/3 + G/3 + B/3)

plt.imshow(grayscale, cmap='gray')
```

```python
# negate grayscale image

grayscale_neg = 255 - grayscale

plt.imshow(grayscale_neg, cmap='gray')
```

```python
# negate color image

image[:,:, 0] = 255 - image[:,:, 0]
image[:,:, 1] = 255 - image[:,:, 1]
image[:,:, 2] = 255 - image[:,:, 2]


plt.imshow(image)
```
