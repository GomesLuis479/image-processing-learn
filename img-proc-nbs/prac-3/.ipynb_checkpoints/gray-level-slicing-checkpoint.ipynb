```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.image as mpimg
```

```python
# read the image

image_original = mpimg.imread(r'../images/dogs.jpg')
image = image_original.copy() # else it is read only

# display original

plt.imshow(image_original, cmap='gray')
```

```python
# converting to grayscale: avg method

R = image[:,:, 0]
G = image[:,:, 1]
B = image[:,:, 2]

grayscale = (R/3 + G/3 + B/3)

plt.imshow(grayscale, cmap='gray')
```

```python
# gray level slicing with background

r1 = int(input('Enter r1: '))
r2 = int(input('Enter r2: '))

GLS_with_bg = np.zeros(grayscale.shape)

for i in range(grayscale.shape[0]):
    for j in range(grayscale.shape[1]):
        t = grayscale[i, j]
        if t >= r1 and t <=r2:
            GLS_with_bg[i, j] = 255
        else:
            GLS_with_bg[i, j] = t
            
plt.imshow(GLS_with_bg, cmap='gray')
```

```python
#gray level slicing without background

r1 = int(input('Enter r1: '))
r2 = int(input('Enter r2: '))

GLS_without_bg = np.zeros(grayscale.shape)

for i in range(grayscale.shape[0]):
    for j in range(grayscale.shape[1]):
        t = grayscale[i, j]
        if t >= r1 and t <=r2:
            GLS_without_bg[i, j] = 255
            
plt.imshow(GLS_without_bg, cmap='gray')
```

```python
# log s = c * log(1+r)

c = int(input('Enter the constant c: '))

log_image = c * np.log10(grayscale + 1)
plt.imshow(log_image, cmap='gray')
```

```python
# power law s = c * r^(gamma)

c = int(input('Enter the constant c: '))
gamma = float(input('Enter gamma: '))

power_image = c * np.power(grayscale, gamma)

plt.imshow(power_image, cmap='gray')
```
