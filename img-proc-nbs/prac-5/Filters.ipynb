```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.image as mpimg
```

```python
# read the image

image_original = mpimg.imread(r'../images/dogs.jpg')
image = image_original.copy() # else it is read only

# display original

plt.imshow(image_original, cmap='gray')
plt.title('Original RGB imamge')
```

```python
# Filtering

# n -> filter size
# filt -> filter kernel

n = int(input('enter the filter size (must be odd): '))
filt = np.zeros((n, n))

print('Enter the values of the filter: ')
for i in range(n):
    for j in range(n):
        temp = float(input('({0}, {1}) : '.format(i, j)))
        filt[i, j] = temp
        
print(filt)
```

```python
# convert to gray

# converting to grayscale: avg method

R = image[:,:, 0]
G = image[:,:, 1]
B = image[:,:, 2]

grayscale = (R/3 + G/3 + B/3)

plt.imshow(grayscale, cmap='gray')
plt.title('grayscale image')
```
