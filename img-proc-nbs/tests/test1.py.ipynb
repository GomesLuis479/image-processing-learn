```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
import matplotlib.image as mpimg
```

```python
# read the image

img = mpimg.imread(r'images/test1.jpg')
image = img.copy() # else it is read only
```

```python
# display image chracteristics

print('This image is: ', type(image), 'with dimensions:', image.shape)
```

```python
print(image)
```

```python
# testing ndArray arrangement

print(image[0][0][1])
```

```python
# display image

plt.imshow(image)
```

```python
# processing 1: boosting the red channel

for row in image:
    for pixel in row:
        inc = pixel[0] + 40
        if inc < 256:
            pixel[0] += 40
        else:
            pixel[0] = 255
```

```python
# display image

plt.imshow(image)
```

```python
def alterImage(image, redShift, greenShift, blueShift):
    
    def alterPixel(pixel, shift, channel):
        newValue = pixel[channel] + shift
        if newValue < 0:
            pixel[channel] = 0
        elif newValue > 255:
            pixel[channel] = 255
        else:
            pixel[channel] += shift
    
    for row in image:
        for pixel in row:
            alterPixel(pixel, redShift, 0)
            alterPixel(pixel, greenShift, 1)
            alterPixel(pixel, blueShift, 2)      
```

```python
image = img.copy()
alterImage(image, 50, 0, 0)
plt.imshow(image)
```
